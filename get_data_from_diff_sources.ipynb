```python
import requests
import sqlite3
import pandas as pd
```

```python
pip install requests
```

Step 1: Load SpaceX Launch Data from API

```python
url = ' https://api.spacexdata.com/v4/launches'
response = requests.get(url)
response
```

```python
launch = response.json()
#Display dataframe
launch  = pd.DataFrame(launch)
launch.head()
```

```python
#Extract relevant columns: name, date_utc, success, details, rocket
launch = launch[['name', 'date_utc', 'success', 'details', 'rocket']]
launch
```

```python
#onvert date_utc to datetime and extract the year
launch['date_utc'] = pd.to_datetime(launch['date_utc'])
launch['year'] = launch['date_utc'].dt.year
launch.head()
```

Step 2: Load Rocket Metadata

```python

url = 'https://api.spacexdata.com/v4/rockets'
response = requests.get(url)
response
```

```python
#Display rocket dataframe
rocket = response.json()
rocket = pd.DataFrame(rocket)
rocket.head()
```

```python
#Extract id, name, type, active, and stages
rocket = rocket[['id', 'name', 'type', 'active', 'stages']]
rocket
```

Step 3: Merge Launch and Rocket Data

```python
#Join the two DataFrames on rocket ID using
launch_rocket = pd.merge(launch, rocket, left_on='rocket', right_on='id')
launch_rocket = launch_rocket.drop('id', axis=1)
launch_rocket.head()
```

Step 4: Add Simulated Country Information

```python
#Add a new column country and randomly
import random
launch_rocket['country'] = [random.choice(['USA', 'Russia', 'India', 'China', 'France']) for _ in range(len(launch_rocket))]
launch_rocket.head()
```

Step 5: Store Merged Data in SQLite3

```python
#create a connection and save the merged DataFrame as a table named launches
conn = sqlite3.connect('spacex.db')
launch_rocket.to_sql('launches', conn, if_exists='replace', index=False)
```

```python
pd.read_sql('SELECT * FROM launches', conn)
```

```python
pd.read_sql('SELECT country, COUNT(*) AS launches FROM launches GROUP BY country', conn)
```

```python
pd.read_sql('SELECT year, COUNT(*) AS launches FROM launches GROUP BY year ORDER BY launches DESC LIMIT 1', conn)
```

```python
pd.read_sql('SELECT name_x, COUNT(*) AS launches FROM launches GROUP BY name_x ORDER BY launches DESC LIMIT 5', conn)
```
